# **BINARY CLASSIFICATION USING KERAS**

In this exercise, we will implement a very simple binary classifier with a single neuron using the Keras API.

Import the required libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# Import Keras libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

Import and inspect the data:

```python
df = pd.read_csv('data/data.csv')
df.head()
```

Visualize the data using a scatter plot:

```python
plt.scatter(df[df['label'] == 0]['x1'], df[df['label'] == 0]['x2'], marker='*')
plt.scatter(df[df['label'] == 1]['x1'], df[df['label'] == 1]['x2'], marker='<');
```

Prepare the data by separating the features and labels and setting the `tf` variables:

```python
X = df[['x1', 'x2']].values
y = df[['label']].values
```

Create a neural network model consisting of a single layer with a neuron and a sigmoid activation function:

```python
model = Sequential()
model.add(Dense(units=1, input_dim=2, activation='sigmoid'))
```

Once the model is created, we can config the model with losses and metrics with `model.compile()`, train the model with `model.fit()`, and use the model to do prediction with `model.predict()`.

```python
# compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# fit the model
model.fit(X, y, epochs=1000)

# evaluate the model
model.evaluate(X, y)
```
